```python
#Import Libraries
import requests
import numpy as np
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
#The list of 19 types of cryptocurrency
coin_list=['BTC','ETH','XRP','ADA','USDT','DOGE','XLM','DOT','UNI','LINK','USDC','BCH','LTC','GRT','ETC','FIL','AAVE','ALGO','EOS']
```

```python
#Defining the dataframe
main_df=pd.DataFrame()
```

```python
api_key="740c713a17980909f11b6f8ceb8d628c0642f29d15991cc3588c3ef9ae97650b"
for coin in coin_list:
    coin_df=pd.DataFrame()
    df=pd.DataFrame(index=[0])

    #Defining time range
    datetime_end=datetime(2025, 2, 25, 0, 0)
    datetime_check=datetime(2025, 2, 15, 0, 0)

    while len(df)>0:
        if datetime_end==datetime_check:
            break
        datetime_start=datetime_end-relativedelta(hours=12)

        #API Request
        url=f"https://data-api.coindesk.com/index/cc/v1/latest/tick?market=ccix&instruments={coin}-USD&api_key={api_key}"
        response=requests.get(url).json()

        #Validate API response
        if "Data" in response and f"{coin}-USD" in response["Data"]:
            data=response["Data"][f"{coin}-USD"]
        else:
            break

        #Extract relevant values
        df=pd.DataFrame(
            {
                "Datetime":[datetime_end],
                "Open":[data.get("CURRENT_HOUR_OPEN",np.nan)],
                "High":[data.get("CURRENT_HOUR_HIGH",np.nan)],
                "Low":[data.get("CURRENT_HOUR_LOW",np.nan)],
                "Close":[data.get("VALUE",np.nan)],
            }
        )

        coin_df=pd.concat([df,coin_df],ignore_index=True)
        datetime_end=datetime_start

    coin_df["Symbol"]=coin
    main_df=pd.concat([main_df,coin_df],ignore_index=True)

```
